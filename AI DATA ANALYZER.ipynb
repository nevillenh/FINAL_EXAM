```python
# Import libraries
import pandas as pd
import numpy as np
# Load the dataset
data = pd.read_csv('teslastockdata.csv')
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Basic statistics
print(data.describe())
```

```python
# DATA CLEANING
# Check for missing values
print(data.isnull().sum())
```

```python
# NOTE: There are NO missing values in the dataset

# To fill missing values -> fillna(data.mean(numeric_only=True), inplace=True)

```

```python

# Check for duplicates
print(data.duplicated().sum())
```

```python
# NOTE: There are NO duplicate rows in the dataset

# To drop duplicates -> data.drop_duplicates(inplace=True)
```

```python
# DATA ANALYSIS
# Grouping
grouped_data = data.groupby('Date')['Close'].mean()
print(grouped_data)
```

```python
# Sorting
sorted_data = data.sort_values(by='Close', ascending=False)
print(sorted_data.head())
```

```python
# Correlation
print(data.corr(numeric_only=True))
```

```python
# DATA VISUALIZATION
# Importing visualization libraries
import matplotlib.pyplot as plt
```

```python
# Histogram
plt.hist(data['Close'], bins=20)
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# line plot
plt.plot(data['Date'], data['Close'])
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

```python
# bar chart
plt.bar(data['Date'], data['Close'])
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```
